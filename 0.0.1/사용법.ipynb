````python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import Tk, ttk, StringVar
import os
import shutil
import subprocess
from datetime import datetime
from glob import glob
import win32com.client
import pickle
import datetime

# 변수
################################################################################################

backup_path = ""
user_name = ""
backup_path2 = ""
delay = ""
cert = ""
selected_value = ""
entered_value = ""

# 함수
################################################################################################

# 백업 후 종료
def backup_and_exit():
    NOW = datetime.datetime.now()
    HOUR = int(NOW.hour)
    MINUTE = int(NOW.minute)
    hour_and_min = entered_value.split(":")
    if (HOUR == int(hour_and_min[0])) and (MINUTE == int(hour_and_min[1])):
        VELOG_BACKUP(backup_path, backup_path2, user_name, delay, cert)
        root.destroy()
        
def delete_backup():
    folder_path = os.path.dirname([os.path.abspath(file) for file in glob("*exe")][0]) + "\\backup"
    if os.path.exists(folder_path):
        subprocess.run(['runas', '/user:Administrator', 'cmd', '/c', 'rmdir', '/s', '/q', folder_path], shell=True, check=True)

def schedule_task():
    # 작업을 수행할 시간 간격 (밀리초 단위)
    interval = 10000  # 1초
    
    # do_something 함수를 호출
    backup_and_exit()

    # 일정 시간이 지난 후에 다시 schedule_task 함수를 호출
    root.after(interval, schedule_task)

# 프로그램 종료, 종료 취소 함수
def exit_program():
    if exit_code == 2:
        root.destroy()

def cancel_exit():
    global exit_code
    global cancel_label
    exit_code = 1
    cancel_label.configure(text="프로그램 종료가 취소되었습니다")
    # 취소된 경우 추가 동작 수행 가능

def start_exit_timer():
    folder_path = os.path.dirname([os.path.abspath(file) for file in glob("*exe")][0]) + "\\backup"
    if os.path.exists(folder_path):
        subprocess.run(['runas', '/user:Administrator', 'cmd', '/c', 'rmdir', '/s', '/q', folder_path], shell=True, check=True)
        messagebox.showinfo("알림", "backup 폴더가 남아있어 삭제되었습니다. 오류가 났다면 재시작해주세요.")
    global exit_timer
    exit_timer = root.after(1800000, exit_program)  # 30분(1800000밀리초) 후에 exit_program 실행

# 데이터 저장 함수
def save_data_to_file(file_path, data):
    with open(file_path, 'wb') as file:
        pickle.dump(data, file)

# 데이터 불러오기 함수
def load_data_from_file(file_path, default_value=None):
    if not os.path.exists(file_path):
        return default_value
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    return data

# 선택란 함수
def get_selected_value():
    # value = combo.get()
    entry_label.pack_forget()
    
# 입력창에 값 입력
def on_entry_enter(event):
    global entered_value
    global selected_value
    entered_value = entry.get()
    selected_value = combo.get()
    if (selected_value == "day"):
        result_label.config(text=f"매일 {entered_value}마다 백업됩니다.")
    else:
        result_label.config(text=f"매주 {selected_value} {entered_value}마다 백업됩니다.")
    exe_path = glob("*exe")

    # 상대경로를 절대경로로 변환
    absolute_paths = [os.path.abspath(file) for file in exe_path]
    register_startup_program(absolute_paths[0])

# 입력창 받아오기
def on_combo_select(*args):
    entry_label.pack_forget()
    entry.pack(before=execute_button)
    result_label.pack(after=entry)

# 시작프로그램 위치 반환 함수
def get_startup_folder():
    startup_folder = subprocess.check_output('echo %APPDATA%\\Microsoft\\Windows\\Start Menu\\Programs\\Startup', shell=True)
    startup_folder = startup_folder.decode().strip()
    return startup_folder

# 시작프로그램의 위치를 받아서 바로가기를 생성하는 함수
def register_startup_program(program_path):
    startup_folder = get_startup_folder()
    shortcut_path = os.path.join(startup_folder, 'velog_backup.lnk')
    
    shell = win32com.client.Dispatch("WScript.Shell")
    shortcut = shell.CreateShortcut(shortcut_path)
    shortcut.TargetPath = program_path
    shortcut.Save()

# 특정 파일이 없을 때 바로가기 삭제하는 함수
def delete_startup_shortcut():
    startup_folder = get_startup_folder()
    shortcut_path = os.path.join(startup_folder, 'velog_backup.lnk')
    
    if os.path.exists(shortcut_path):
        os.remove(shortcut_path)
        
# 설정 초기화 함수
def delete_txt():
    if os.path.exists('init_setting.txt'):
        os.remove('init_setting.txt')
    startup_folder = get_startup_folder()
    shortcut_path = os.path.join(startup_folder, 'velog_backup.lnk')
    
    if os.path.exists(shortcut_path):
        os.remove(shortcut_path)

# 폴더 생성 함수
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
def VELOG_BACKUP(VelogBackup, MOVE_PATH, USER_NAME, DELAY, CERT):
    
    if MOVE_PATH == "":
        MOVE_PATH = VelogBackup
        
    # ROOT 폴더 경로
    ROOT_PATH = f"backup"

    # 대상 폴더 경로
    CONTENT_PATH = f"{ROOT_PATH}/content"

    # 이동 대상 폴더 경로
    SOURCE_PATH = f"{ROOT_PATH}/images"
    
    # 이동할 새 경로
    NEW_PATH = MOVE_PATH + f"/{USER_NAME}_velog_backup"

    # 기존 backup 폴더 삭제
    if os.path.exists(ROOT_PATH):
        shutil.rmtree(ROOT_PATH)
    
    if os.path.exists(NEW_PATH + "/backup"):
        shutil.rmtree(NEW_PATH + "/backup")
    if os.path.exists(NEW_PATH + "/content"):
        shutil.rmtree(NEW_PATH + "/content")
    if os.path.exists(NEW_PATH + "/images"):
        shutil.rmtree(NEW_PATH + "/images")
    
    if DELAY != "":
        DELAY = f"-d {DELAY}"
        
    if CERT != "":
        CERT = f"-c {CERT}"

    # cmd 명령어 실행
    subprocess.run(['cmd', '/c', f'node {VelogBackup}/app.js -u {USER_NAME} {DELAY} {CERT}'])

    # 파일 이동
    shutil.move(SOURCE_PATH, CONTENT_PATH)

    # 대상 폴더 내의 모든 파일 경로 탐색

    MD_LIST = glob(f'{CONTENT_PATH}/*.md')

    for file_name in MD_LIST:
        # 파일 열기
        with open(file_name, 'r', encoding='UTF-8') as file:
            content = file.read()
            
        # 이미지 경로 변경
        new_content = content.replace(f'/images/', 'images/')
        
        # 변경된 내용 저장
        with open(file_name, 'w', encoding = 'UTF-8') as file:
            file.write(new_content)

    # backup/content 폴더 이름을 backup-유저명-날짜로 변경
    
    # 이동할 새 경로
    createFolder(NEW_PATH)

    # 폴더 이동
    shutil.move(CONTENT_PATH, NEW_PATH)

    # 폴더 이름 변경
    os.rename(os.path.join(NEW_PATH, 'content'), os.path.join(NEW_PATH, f'velog_backup_{datetime.datetime.today().strftime("%Y%m%d_%H%M%S")}'))
    
    # backup 폴더 삭제
    shutil.rmtree(ROOT_PATH)

# velog_backup app.js 경로
def select_path():
    path = filedialog.askdirectory()
    path_label.configure(text=path)
    global backup_path
    backup_path = path_label.cget("text")

# 백업 폴더가 저장될 경로
def select_path2():
    path2 = filedialog.askdirectory()
    path_label2.configure(text=path2)
    global backup_path2
    backup_path2 = path_label2.cget("text")

# 유저명 가져오기
def get_username(*args):
    text3 = username_input.get()
    username_label.configure(text=text3)
    global user_name
    user_name = username_label.cget("text")
    
# 딜레이 가져오기
def get_delay(*args):
    text4 = delay_input.get()
    delay_label.configure(text=text4)
    global delay
    delay = delay_label.cget("text")
    
# 엑세스 토큰 가져오기
def get_access_token(*args):
    text5 = access_token_input.get()
    access_token_label.configure(text=text5)
    global cert
    cert = access_token_label.cget("text")

# 작업이 끝나면 알림
def execute():
    # 실행할 코드 작성
    messagebox.showinfo("Velog Backup", f"백업이 {backup_path2}/{user_name}_velog_backup에 저장되었습니다.")

# 프로그램 시작시 체크
################################################################################################

# 프로그램 시작 시 특정 파일 유무에 따라 바로가기 생성 또는 삭제
if os.path.exists('init_setting.txt'):
    txt_path = glob("*txt")
    # 상대경로를 절대경로로 변환
    txt_absolute_paths = [os.path.abspath(file) for file in txt_path]
    data_list = load_data_from_file(txt_absolute_paths[0])
    backup_path = data_list[0]
    user_name = data_list[1]
    backup_path2 = data_list[2]
    delay = data_list[3]
    cert = data_list[4]
    selected_value = data_list[5]
    entered_value = data_list[6]
else:
    delete_startup_shortcut()
    
if selected_value == "" or entered_value == "":
    delete_startup_shortcut()

# twinker GUI 세팅
################################################################################################
root = tk.Tk()

# root.withdraw()  # 윈도우 숨기기
# root.iconify()  # 윈도우 최소화

exit_code = 2
# 타이머
start_exit_timer()

# 초기 작업 스케줄링
NOW = datetime.datetime.now()
today = NOW.strftime('%A')
if today == selected_value:
    schedule_task()

# 창 크기 설정
root.geometry("300x550")

# 창 이름 설정
root.title(f"Velog Backup")

# 경로 선택 창
path_button1 = tk.Button(root, text="Select velog-backup Path \n(app.js 파일이 들어있는 폴더를 선택해주세요.)", command=select_path)
path_label = tk.Label(root, text=backup_path)
path_button1.pack()
path_label.pack()

# 텍스트 입력 창 1
username_input = tk.Entry(root)
username_input.pack()
username_button = tk.Button(root, text="유저명 (필수 입력)", command=get_username)
username_label = tk.Label(root, text=user_name)
username_button.pack()
username_label.pack()

# 텍스트 입력 창 2
delay_input = tk.Entry(root)
delay_input.pack()
delay_button = tk.Button(root, text="딜레이 (단위 ms, 글이 많은 경우 늘려주세요.\n기본값은 0입니다.)", command=get_delay)
delay_label = tk.Label(root, text=delay)
delay_button.pack()
delay_label.pack()

# 텍스트 입력 창 3
access_token_input = tk.Entry(root)
access_token_input.pack()
access_token_button = tk.Button(root, text="velog 유저 엑세스 토큰 \n(필요시 사용, 기본값은 공백입니다.)", command=get_access_token)
access_token_label = tk.Label(root, text=cert)
access_token_button.pack()
access_token_label.pack()

# 경로 선택 창
path_button2 = tk.Button(root, text="백업 폴더를 생성할 위치 \n(유저명_velog_backup으로 폴더가 생성됩니다.)", command=select_path2)
path_label2 = tk.Label(root, text=backup_path2)
path_button2.pack()
path_label2.pack()

# ComboBox 생성
combo = ttk.Combobox(root, values=["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"])
combo.pack()

# 선택창
entry = ttk.Entry(root)
entry.pack_forget()

# 입력창에 이벤트 바인딩
entry.bind("<Return>", on_entry_enter)
entry_label = tk.Label(root, text=selected_value + " "+  entered_value)
entry_label.pack()

# 확인 버튼 생성 # get_selected_value, show entry
# button = ttk.Button(root, text="확인", command=confirm) # 시작프로그램에 바로가기 생성
# button.pack()

# ComboBox 값 변경 시 콜백 함수 호출
combo_var = StringVar()
combo_var.trace("w", on_combo_select)
combo.config(textvariable=combo_var)

# 텍스트 보여주기
result_label = tk.Label(root, text=selected_value + " " +  entered_value)
result_label.pack_forget()

# 실행 버튼
execute_button = tk.Button(root, text="실행", command=lambda : (delete_backup(), VELOG_BACKUP(backup_path, backup_path2, user_name, delay, cert), save_data_to_file(os.path.dirname([os.path.abspath(file) for file in glob("*exe")][0]) + "\\init_setting.txt", [backup_path, user_name, backup_path2, delay, cert, selected_value, entered_value]), execute()))
execute_button.pack()

# 설정 초기화
button = ttk.Button(root, text="설정 초기화", command=delete_txt) # 시작프로그램에 바로가기 생성
button.pack()

cancel_button = tk.Button(root, text="프로그램 종료 취소 \n(30분 뒤에 프로그램이 자동 종료됩니다.)", command=cancel_exit)
cancel_button.pack()
cancel_label = tk.Label(root, text="")
cancel_label.pack()

# GUI 창 실행
root.mainloop()
````

````python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
from tkinter import ttk
import os
import shutil
import subprocess
from datetime import datetime
from glob import glob
import pickle
import datetime

# 변수
################################################################################################

backup_path = ""
user_name = ""
backup_path2 = ""
delay = ""
cert = ""

# 함수
################################################################################################

def delete_backup():
    folder_path = os.path.dirname([os.path.abspath(file) for file in glob("*exe")][0]) + "\\backup"
    if os.path.exists(folder_path):
        subprocess.run(['cmd', '/c', 'rmdir', '/s', '/q', folder_path], shell=True, check=True)
        messagebox.showinfo("알림", "backup 폴더가 남아있어 삭제되었습니다. 오류가 발생한다면 재시작해주세요.")

# 데이터 저장 함수
def save_data_to_file(file_path, data):
    with open(file_path, 'wb') as file:
        pickle.dump(data, file)

# 데이터 불러오기 함수
def load_data_from_file(file_path, default_value=None):
    if not os.path.exists(file_path):
        return default_value
    with open(file_path, 'rb') as file:
        data = pickle.load(file)
    return data
        
# 설정 초기화 함수
def delete_txt():
    if os.path.exists('init_setting.txt'):
        os.remove('init_setting.txt')

# 폴더 생성 함수
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        
def VELOG_BACKUP(VelogBackup, MOVE_PATH, USER_NAME, DELAY, CERT):
    
    if MOVE_PATH == "":
        MOVE_PATH = VelogBackup
        
    # ROOT 폴더 경로
    ROOT_PATH = f"{backup_path}/backup"

    # 대상 폴더 경로
    CONTENT_PATH = f"{ROOT_PATH}/content"

    # 이동 대상 폴더 경로
    SOURCE_PATH = f"{ROOT_PATH}/images"
    
    # 이동할 새 경로
    NEW_PATH = MOVE_PATH + f"/{USER_NAME}_velog_backup"

    # 기존 backup 폴더 삭제
    if os.path.exists(ROOT_PATH):
        shutil.rmtree(ROOT_PATH)
    
    if os.path.exists(NEW_PATH + "/backup"):
        shutil.rmtree(NEW_PATH + "/backup")
    if os.path.exists(NEW_PATH + "/content"):
        shutil.rmtree(NEW_PATH + "/content")
    if os.path.exists(NEW_PATH + "/images"):
        shutil.rmtree(NEW_PATH + "/images")
    
    if DELAY != "":
        DELAY = f"-d {DELAY}"
        
    if CERT != "":
        CERT = f"-c {CERT}"

    # cmd 명령어 실행
    subprocess.run(['cmd', '/c', f'node {VelogBackup}/app.js -u {USER_NAME} {DELAY} {CERT}'])

    # 파일 이동
    shutil.move(SOURCE_PATH, CONTENT_PATH)

    # 대상 폴더 내의 모든 파일 경로 탐색

    MD_LIST = glob(f'{CONTENT_PATH}/*.md')

    for file_name in MD_LIST:
        # 파일 열기
        with open(file_name, 'r', encoding='UTF-8') as file:
            content = file.read()
            
        # 이미지 경로 변경
        new_content = content.replace(f'/images/', 'images/')
        
        # 변경된 내용 저장
        with open(file_name, 'w', encoding = 'UTF-8') as file:
            file.write(new_content)

    # backup/content 폴더 이름을 backup-유저명-날짜로 변경
    
    # 이동할 새 경로
    createFolder(NEW_PATH)

    # 폴더 이동
    shutil.move(CONTENT_PATH, NEW_PATH)

    # 폴더 이름 변경
    os.rename(os.path.join(NEW_PATH, 'content'), os.path.join(NEW_PATH, f'velog_backup_{datetime.datetime.today().strftime("%Y%m%d_%H%M%S")}'))
    
    # backup 폴더 삭제
    shutil.rmtree(ROOT_PATH)

# velog_backup app.js 경로
def select_path():
    path = filedialog.askdirectory()
    path_label.configure(text=path)
    global backup_path
    backup_path = path_label.cget("text")

# 백업 폴더가 저장될 경로
def select_path2():
    path2 = filedialog.askdirectory()
    path_label2.configure(text=path2)
    global backup_path2
    backup_path2 = path_label2.cget("text")

# 유저명 가져오기
def get_username(*args):
    text3 = username_input.get()
    username_label.configure(text=text3)
    global user_name
    user_name = username_label.cget("text")
    
# 딜레이 가져오기
def get_delay(*args):
    text4 = delay_input.get()
    delay_label.configure(text=text4)
    global delay
    delay = delay_label.cget("text")
    
# 엑세스 토큰 가져오기
def get_access_token(*args):
    text5 = access_token_input.get()
    access_token_label.configure(text=text5)
    global cert
    cert = access_token_label.cget("text")

# 작업이 끝나면 알림
def execute():
    # 실행할 코드 작성
    messagebox.showinfo("Velog Backup", f"백업이 {backup_path2}/{user_name}_velog_backup에 저장되었습니다.")

# 프로그램 시작시 체크
################################################################################################

# 프로그램 시작 시 특정 파일 유무에 따라 바로가기 생성 또는 삭제
if os.path.exists('init_setting.txt'):
    txt_path = glob("*txt")
    # 상대경로를 절대경로로 변환
    txt_absolute_paths = [os.path.abspath(file) for file in txt_path]
    data_list = load_data_from_file(txt_absolute_paths[0])
    backup_path = data_list[0]
    user_name = data_list[1]
    backup_path2 = data_list[2]
    delay = data_list[3]
    cert = data_list[4]

# twinker GUI 세팅
################################################################################################
root = tk.Tk()

# 창 크기 설정
root.geometry("300x450")

# 창 이름 설정
root.title(f"Velog Backup")

# 경로 선택 창
path_button1 = tk.Button(root, text="Select velog-backup Path \n(app.js 파일이 들어있는 폴더를 선택해주세요.)", command=select_path)
path_label = tk.Label(root, text=backup_path)
path_button1.pack()
path_label.pack()

# 텍스트 입력 창 1
username_input = tk.Entry(root)
username_input.pack()
username_button = tk.Button(root, text="유저명 (필수 입력)", command=get_username)
username_label = tk.Label(root, text=user_name)
username_button.pack()
username_label.pack()

# 텍스트 입력 창 2
delay_input = tk.Entry(root)
delay_input.pack()
delay_button = tk.Button(root, text="딜레이 (단위 ms, 글이 많은 경우 늘려주세요.\n기본값은 0입니다.)", command=get_delay)
delay_label = tk.Label(root, text=delay)
delay_button.pack()
delay_label.pack()

# 텍스트 입력 창 3
access_token_input = tk.Entry(root)
access_token_input.pack()
access_token_button = tk.Button(root, text="velog 유저 엑세스 토큰 \n(필요시 사용, 기본값은 공백입니다.)", command=get_access_token)
access_token_label = tk.Label(root, text=cert)
access_token_button.pack()
access_token_label.pack()

# 경로 선택 창
path_button2 = tk.Button(root, text="백업 폴더를 생성할 위치 \n(유저명_velog_backup으로 폴더가 생성됩니다.)", command=select_path2)
path_label2 = tk.Label(root, text=backup_path2)
path_button2.pack()
path_label2.pack()

# 실행 버튼
execute_button = tk.Button(root, text="실행", command=lambda : (delete_backup(), VELOG_BACKUP(backup_path, backup_path2, user_name, delay, cert), save_data_to_file(os.path.dirname([os.path.abspath(file) for file in glob("*exe")][0]) + "\\init_setting.txt", [backup_path, user_name, backup_path2, delay, cert]), execute()))
execute_button.pack()

# 설정 초기화
button = ttk.Button(root, text="설정 초기화", command=delete_txt) # 시작프로그램에 바로가기 생성
button.pack(pady=20)

# GUI 창 실행
root.mainloop()
````
